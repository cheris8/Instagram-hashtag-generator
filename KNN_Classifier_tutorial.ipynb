```python
import KNN_Classifier
from keras.preprocessing import image
```

```python
#클래스 선언
knn = KNN_Classifier.KNN_Classifier()
```

```python
#target image 설정, 이를 위해 from keras.preprocessing import image 를 임포트 해주어야합니다.
target = image.load_img('data/total_img/데일리_36.jpg',target_size = (224,224))
```

```python
#target image를 보여줍니다.
knn.show_target(target)
```

```python
#knn.make_picke() 이건 data.pickle 파일이 없을 때 사용하시면 됩니다. pickle 까지 git 에 올려드렸어요
```

```python
#이미 저장되어 있는 data.pickle 을 불러옵니다.
knn.load_pickle()
```

```python
#target image 를 전처리 해주어 유사도를 구할 수 있도록 해줍니다.
tar = knn.img_preprocess(target)
tar = knn.model(tar)
```

```python
#이 함수까지는 필수로 호출하셔야 hashtag_list 를 뽑아올 수 있어요
#두 번째 인자는, 유사도 상위 몇개까지 불러올 것인가입니다. 보통 5가 좋아요
knn.show_KNN(tar,5)
```

```python
### KNN ###

import KNN_Classifier

# 클래스 선언
knn = KNN_Classifier.KNN_Classifier()

# 모델 불러오기
knn.load_pickle()

# 타겟 이미지 설정
from keras.preprocessing import image
target = image.load_img('../data/img/데일리_36.jpg',target_size = (224,224))

# 타겟 이미지 전처리
tar = knn.img_preprocess(target)
tar = knn.model(tar)

# 유사도 기준 상위 5개 출력
knn.show_KNN(tar, 5)
```

```python
### 토큰 전환 ###

# 공통 토큰 불러오기
common_tokens = knn.get_comm_token(2)

# Word2Vec 모델 불러오기
from gensim.models.word2vec import Word2Vec
model = Word2Vec.load('Word2Vec_model')

# 토큰 전환 실행하기
from TokenConversion import FirstTokenConversion, NTokenConversion, TokenConversion
tokens_converted = TokenConversion(model, common_tokens, alpha=0.7)
tokens_converted
```
